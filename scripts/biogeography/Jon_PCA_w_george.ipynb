```r
library("RcppCNPy")
library("ggplot2")
library(ggrepel)
library(dplyr)
library(tidyr)
library(ggpubr)
```

```r
both.e1 <- as.matrix(read.table("Vanderbilt/Updated/jon.tank.george.Q20.admixe1.cov"))
both.e2 <- as.matrix(read.table("Vanderbilt/Updated/jon.tank.george.Q20.admixe2.cov"))
both.e3 <- as.matrix(read.table("Vanderbilt/Updated/jon.tank.george.Q20.admixe3.cov"))
both.e4 <- as.matrix(read.table("Vanderbilt/Updated/jon.tank.george.Q20.admixe4.cov"))
```

```r
location <- read.table("Vanderbilt/jon.tank.george.run.tab", sep="\t", header=T)
```

```r
e_both.e1 <- eigen(both.e1)$vectors
colnames(e_both.e1) <- colnames(both.e1)
e_both.e2 <- eigen(both.e2)$vectors
colnames(e_both.e2) <- colnames(both.e2)
e_both.e3 <- eigen(both.e3)$vectors
colnames(e_both.e3) <- colnames(both.e3)
e_both.e4 <- eigen(both.e4)$vectors
colnames(e_both.e4) <- colnames(both.e4)
```

```r
both_df.e1 <- data.frame(PC1=as.data.frame(e_both.e1)$V1, 
                     PC2=as.data.frame(e_both.e1)$V2, Sample=location$Sample, Location=location$Location, 
                            SampleName=location$Sample)
both_df.e2 <- data.frame(PC1=as.data.frame(e_both.e2)$V1, 
                     PC2=as.data.frame(e_both.e2)$V2, Sample=location$Sample, Location=location$Location, 
                            SampleName=location$Sample)
both_df.e3 <- data.frame(PC1=as.data.frame(e_both.e3)$V1, 
                     PC2=as.data.frame(e_both.e3)$V2, Sample=location$Sample, Location=location$Location, 
                            SampleName=location$Sample)
both_df.e4 <- data.frame(PC1=as.data.frame(e_both.e4)$V1, 
                     PC2=as.data.frame(e_both.e4)$V2, Sample=location$Sample, Location=location$Location, 
                            SampleName=location$Sample)
```

```r
options(repr.plot.width = 12, repr.plot.height = 8)
```

```r
e2 <- ggplot(both_df.e2, aes(x=PC1, y=PC2, size=1.5, color=Location, label=Location))  + geom_point() + geom_label_repel(max.overlaps=20) 
```

```r
e3 <- ggplot(both_df.e3, aes(x=PC1, y=PC2, size=1.5, color=Location, label=Location))  + geom_point() + geom_label_repel(max.overlaps=20) 
```

```r
e4 <- ggplot(both_df.e4, aes(x=PC1, y=PC2, size=1.5, color=Location, label=Location))  + geom_point() + geom_label_repel(max.overlaps=20) 
```

```r
e1 <- ggplot(both_df.e1, aes(x=PC1, y=PC2, size=1.5, color=Location, label=Location))  + geom_point() + geom_label_repel(max.overlaps=20) 
```

```r
options(repr.plot.width = 18, repr.plot.height = 18)
```

```r
ggarrange(e1, e2, e3, e4, labels=c("e=1", "e=2", "e=3", "e=4"))
```
